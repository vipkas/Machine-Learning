## Definition

Decision tree is one of supervised machine learning model. Decision tree is a graph which is used to model a decision. It represents a decision making concept visually and explicitly. Decision tree is also known as CART (Classification and Regression Trees) which is introduced by Leo Breiman to point an algorithm that used for classification and regression. CART algorithm provides some important techniques such **bagged decision trees, random forest and boosted decision trees**.

![features example](assets/pic1.png)

Decision tree model is binary tree which consists of several components, such:

* Each **Node** represents data features (attributes/ x variables) [see Figure 7.1].
* Each **Branch** or **Link** or **Edge** represents the decision or rule.
* Each **leaf** represents decision result or variable output that is used as prediction.

Some terms are also applied for decision tree, such:

* **Splitting**: This kind of process to split a node becomes two or more sub-node.
* **Decision Node**: A form of node when processed to be following sub-node.
* **Pruning**: Deleting a sub-node. 
* **Branch / Sub-Tree**: A part of tree.
* **Parent and Child Node (Node Induk dan Anak)**: Node, which is divided into several sub-node. The node which is on the upper level will be the parent and the following ones will be the child one.

## Tahapan Algoritma Decision Tree

In Decision tree, creating tree is the first step that must be conducted. You may use Information Gain or Gini Index to determine the Root node and its following node. This is the tree creating steps using top-down scheme:


1. At first iteration, a Root Node is choosen based on the lowest value of Gini Index or the highest value of Information Gain of all values in features (x variable) in training data.
2. Try to separate S set to produce data subset.
3. Then, repeat this processes to determine the next decision node. this determination also uses the lowest value of Gini Index or the highest value of Information Gain, but not includes non-relevant previous features.
4. This Algorithm will repeat until all data subset find the leaf that represents class label.

### Training Process - classification tree
In this section, you will perform data tarining using [Wisconsin Breast Cancer Dataset](https://www.kaggle.com/uciml/breast-cancer-wisconsin-data) UCI machine learning repository. You will practice a learning of cancer varians prediction using two features: cancer average radius (```radius_mean```) and cancer average of concave point (```concave points_mean```).

### Preprocess

````python
wbc = pd.read_csv('./dataset/wbc.csv')
wbc.head()
````

````python
X = wbc[['radius_mean', 'concave points_mean']]
y = wbc['diagnosis']
y = y.map({'M':1, 'B':0})
````

````python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
````

````python
from sklearn.tree import DecisionTreeClassifier

# Instantiate DecisionTreeClassifier 'dt' dengan kedalaman maksimum 6
dt = DecisionTreeClassifier(max_depth=6, random_state=1)

# Sesuaikan dt ke set training
dt.fit(X_train, y_train)

# Pmemprediksi label set test
y_pred = dt.predict(X_test)
print(y_pred[0:5])
````

### Classification tree evaluation

````python
from sklearn.metrics import accuracy_score

# Memprediksi test set labels
y_pred = dt.predict(X_test)

#  menghitung set accuracy
acc = accuracy_score(y_test, y_pred)
print("Test set accuracy: {:.2f}".format(acc))
````

## Classification tree Learning
- Explanation of block diagram
    - Decision-Tree: data structure which consists of node hierarchical
    - Node: question or prediction
    - Three kind of nodes
        - **Root**: no node parent, question produces two children nodes.
        - **Internal node**: only have one node parent, question produces two children nodes.
        - **Leaf**: only one node parent, it has no node children.
- Information Gain (IG)
![information gain](https://github.com/goodboychan/chans_jupyter/blob/master/_notebooks/image/ig.png?raw=1)
$$ IG(\underbrace{f}_{\text{feature}}, \underbrace{sp}_{\text{split-point}} ) = I(\text{parent}) - \big( \frac{N_{\text{left}}}{N}I(\text{left}) + \frac{N_{\text{right}}}{N}I(\text{right})  \big) $$
    - It is the criterion to measure the impurity of the node $I(\text{node})$:
        - gini index
        - entropy
        - dst...
- Classification-Tree learning 
    - Node grows recursively.
    - In every node, separate the data based on:
        - feature $f$ and split-point $sp$ to maximize $IG(\text{node})$.
        - id $IG(\text{node}) = 0$, declare the leaf node

### Using entrophy as criterion


````python
from sklearn.tree import DecisionTreeClassifier

# Instantiate dt_entropy, set 'entropy' sebagai kriteria informasi
dt_entropy = DecisionTreeClassifier(max_depth=8, criterion='entropy', random_state=1)

# Fit dt_entropy kedalam training set
dt_entropy.fit(X_train, y_train)
````

### Entropy vs Gini index
In this practice, you will compare accuration between  dt_entropy with another tree named ```dt_gini```. A ```dt_gini```is trained using same parameters except information criterion using keyword ```'gini'```.

````python
dt_gini = DecisionTreeClassifier(max_depth=8, criterion='gini', random_state=1)
dt_gini.fit(X_train, y_train)
````

````python
from sklearn.metrics import accuracy_score

# menggunakan dt_entropy untuk memprediksi test set labels
y_pred = dt_entropy.predict(X_test)
y_pred_gini = dt_gini.predict(X_test)

# mengevaluasi accuracy_entropy
accuracy_entropy = accuracy_score(y_test, y_pred)
accuracy_gini = accuracy_score(y_test, y_pred_gini)

# Print accuracy_entropy
print("Accuracy achieved by using entropy: ", accuracy_entropy)

# Print accuracy_gini
print("Accuracy achieved by using gini: ", accuracy_gini)
````

## Decision tree for regression
- Information criterion fro Regression Tree
$$ I(\text{node}) = \underbrace{\text{MSE}(\text{node})}_{\text{mean-squared-error}} = \dfrac{1}{N_{\text{node}}} \sum_{i \in \text{node}} \big(y^{(i)} - \hat{y}_{\text{node}}  \big)^2 $$
$$ \underbrace{\hat{y}_{\text{node}}}_{\text{mean-target-value}} = \dfrac{1}{N_{\text{node}}} \sum_{i \in \text{node}}y^{(i)}$$
- Prediction
$$ \hat{y}_{\text{pred}}(\text{leaf}) = \dfrac{1}{N_{\text{leaf}}} \sum_{i \in \text{leaf}} y^{(i)}$$

### Training using regression tree
In this practice, you will use regression tree to predict fuel consumption in *mpg* (miles per gallon) using [auto-mpg dataset](https://www.kaggle.com/uciml/autompg-dataset). It contains six features.



#### Preprocess

````python
mpg = pd.read_csv('./dataset/auto-mpg.csv')
mpg.head()
````

````python
mpg = pd.get_dummies(mpg)
````

````python
mpg.head()
````

````python
X = mpg.drop('mpg', axis='columns')
y = mpg['mpg']
````

````python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=3)
````

````python
from sklearn.tree import DecisionTreeRegressor

# Instantiate dt
dt = DecisionTreeRegressor(max_depth=8, min_samples_leaf=0.13, random_state=3)

# Fit dt kedalam the training set
dt.fit(X_train, y_train)
````

### Regression tree evaluation
In this practice, you will evaluate ```dt``` set test perform using *Root Mean Squared Error* (RMSE) metric. RMSE model is used to meausre the average, explicitly how many rate of model prediction that is produced from actual different label. RMSE model can be obtained by measure square root of Mean Squared Error (MSE).



````python
from sklearn.metrics import mean_squared_error

# Compute y_pred
y_pred = dt.predict(X_test)

# Compute mse_dt
mse_dt = mean_squared_error(y_test, y_pred)

# Compute rmse_dt
rmse_dt = mse_dt ** (1/2)

# Print rmse_dt
print("Test set RMSE dt: {:.2f}".format(rmse_dt))
````

### Linear regression vs regression tree
In this practice, you will compare RMSE test set from ```dt``` based on what you can get from linear regression model. You also make an instance from linear regression model ```lr``` the train it using same dataset with ```dt```.



#### Preprocess

````python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()

lr.fit(X_train, y_train)
````

````python
# Predict test set labels
y_pred_lr = lr.predict(X_test)

# Compute mse_lr
mse_lr = mean_squared_error(y_test, y_pred_lr)

# Compute rmse_lr
rmse_lr = mse_lr ** 0.5

# Print rmse_lr
print("Linear Regression test set RMSE: {:.2f}".format(rmse_lr))

# Print rmse_dt
print("Regression Tree test set RMSE: {:.2f}".format(rmse_dt))
````
